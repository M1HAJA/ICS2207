```python
#print keyboard layout
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont

keyboard = ['k', 'f', 'l', 'r', 'h', 'w', 'o', 'a', 'z', 'j', ';', 'u', 't', 'c', 'i', 'e', 's', 'y', ',', 'q', 'x', 'd', '?', 'g', 'n', 'v', 'm', 'p', '.', 'b']
keyCoord = {0:(40,30), 1:(100,30), 2:(165,30), 3:(225,30), 4:(290,30), 5:(350,30), 6:(410,30), 7:(475,30), 8:(535,30), 9:(595,30), 10:(50,95), 11:(110,95), 12:(175,95), 13:(235,95), 14:(295,95), 15:(360,95), 16:(425,95), 17:(485,95), 18:(550,95), 19:(610,95), 20:(80,160), 21:(145,160), 22:(205,160), 23:(270,160), 24:(330,160), 25:(390,160), 26:(450,160), 27:(520,160), 28:(580,160), 29:(640,160)}
        
#print(keyCoord)
    
#opens empty keyboard layout template image
img = Image.open("keyboard.png")

Drawer = ImageDraw.Draw(img)

font = ImageFont.truetype("calibri.ttf", 26)

#adding text to keyboard image
for i in range(0, len(keyboard)):
    Drawer.text(keyCoord[i], keyboard[i], font=font, fill=(0,0,0))

display(img)

#saving result to file
img.save("TestCase7.png")
```
